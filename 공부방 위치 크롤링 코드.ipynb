# 경쟁사 공부방 위치 크롤링

```python
driver = webdriver.Edge()
keyword = '푸르넷공부방'
kakao_url = f"https://map.kakao.com/?q={keyword}"

driver.get(kakao_url)
driver.get(kakao_url)
```

```python
ind = 1  # 현재 복사한 순서
no = 1  # 1~5페이지 중 위치한 곳
page = 1  # 현재 페이지 번호
list2 = []  # 결과물이 저장되는 리스트

while 1:
    try:
        # 업체명, 주소
        title = driver.find_element(by='xpath',
                                    value = f'//*[@id="info.search.place.list"]/li[{ind}]/div[3]/strong/a[2]').text # 해당되는 것을 텍스트로 반환
        addr = driver.find_element(by='xpath',
                                   value = f'//*[@id="info.search.place.list"]/li[{ind}]/div[5]/div[2]/p[1]').text
        add_a = driver.find_element(by='xpath',
                                   value = f'//*[@id="info.search.place.list"]/li[{ind}]/div[5]/div[2]/p[2]').text
        
        list2.append([title, addr, add_a]) #리스트 마지막에 데이터 추가
        
        ind += 1
        
    except:
        # 더보기 버튼 찾기
        if driver.find_element(by='xpath',value = f'//*[@id="info.search.place.more"]').is_displayed(): # 해당되는 값이 있으면 True로 반환
            driver.find_element(by='xpath',value = f'//*[@id="info.search.place.more"]').click()
            no += 1
            ind = 1
            page += 1
            continue

        # 5페이지 단위마다 다음 페이지 버튼 누르기# 다음 페이지로 이동
        elif no >= 5:
            driver.find_element(by='xpath',value = f'//*[@id="info.search.page.next"]').click()
            no = 1
            ind = 1
            page += 1
            continue
            
        # 다음 페이지로 이동
        elif driver.find_element(by='xpath',value = f'//*[@id="info.search.page.no{no+1}"]').is_displayed():
            no += 1
            driver.find_element(by='xpath',value = f'//*[@id="info.search.page.no{no}"]').click()
            ind = 1
            page += 1
            continue
        
        # 더이상 이동할수 없을 경우 종료
        else:
            break
```

```python
pd.DataFrame(list2)
```
